```python
import pandas as pd
import numpy as np
```

### Load the dataset

```python
df = pd.read_csv('iris.csv')
```

```python
df
```

### Group by species

```python
df.groupby(['species'])
df
```

### Descriptive stats for the groups

```python
df.describe()
df.groupby('species').describe()
```

### Ranking the petal length in an ascending order

```python
df['rank'] = df['petal_length'].rank()
df.sort_values('rank', ascending=True)
```

### Mean of each column for each species

```python
df.groupby(['species']).mean()
```

### DataFrame for each of the species separately, with each index starting at 0

```python
df_setosa = df.groupby('species').get_group('setosa')
df_setosa
```

```python
df_virginica = df.groupby('species').get_group('virginica').reset_index()
df_virginica
```

```python
df_versicolor = df.groupby('species').get_group('versicolor').reset_index()
df_versicolor
```

### Two seperate dataframes, one for the Sepal information and one for the Petal information

```python
df_sepal_info = df.filter(['sepal_length', 'sepal_width', 'species'], axis=1)
df_sepal_info
```
